```python
# checking missing values
df.isnull().sum()
```

```python
# how should we replace the null values the answer we should say mean or median
# why mean or why median?
# Generally we should choose median over mean because mean can get affected by outliers and that would skew the distibution of our data
# and why median? Because median is robust to outliers
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#feature engneering
#create a column age_group
labels=['Young Adult', 'Adult','Middle-aged', 'Senior']
df['age_group']=pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quaterly':90,
    'Bi-Weekly':15,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
